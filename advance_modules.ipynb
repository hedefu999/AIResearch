````python
# 随机游走
from random import choice
import matplotlib.pyplot as plt
class RandomWalk:
    """
    生成随机游走数的类
    """
    def __init__(self, num_points=5000):
        self.num_points = num_points
        # 起始位置
        self.x_values = [0]
        self.y_values = [0]

    def get_step(self):
        direction=choice([-1,1])
        distance=choice([1,2,4])
        return direction*distance

    def fill_walk(self):
        while len(self.x_values) < self.num_points:
            x_step=self.get_step()
            y_step=self.get_step()

            # python支持负索引，-1表示最后一个元素，不必像java那样len()-1方法来获取了
            x=self.x_values[-1]+x_step
            y=self.y_values[-1]+y_step

            self.x_values.append(x)
            self.y_values.append(y)
if __name__  == '__main__':
    rw=RandomWalk()
    rw.fill_walk()
    plt.style.use('dark_background')
    fig,ax = plt.subplots(figsize=(20,20))
    point_numbers = range(rw.num_points)
    # 这种颜色方案可以显示随机点的走向
    ax.scatter(rw.x_values, rw.y_values, c=point_numbers,cmap=plt.cm.Blues,edgecolors='none',s=15)
    # 设置 x y 轴上的刻度间距相等
    ax.set_aspect('equal')
    # 绘制起点和终点
    ax.scatter(0,0,c='green',s=100)
    ax.scatter(rw.x_values[-1],rw.y_values[-1],c='red',s=100)
    # 隐藏坐标轴
    ax.get_xaxis().set_visible(False)
    ax.get_yaxis().set_visible(False)
    plt.show()
````

````python
# 这次来掷筛子
from random import randint
import plotly.express as px

class Die:
    def __init__(self, num_sides=6):
        self.num_sides = num_sides
    def roll(self):
        return randint(1, self.num_sides)

die=Die()
results = [die.roll() for x in range(1000)]
print(results)
frequencies = [results.count(value) for value in range(1,die.num_sides+1)]
print(frequencies)

# 使用plotly绘制直方图
title="掷骰子的分布直方图"
labels={'x':'Result','y':'出现频率'}
fig=px.bar(x=[1,2,3,4,5,6],y=frequencies,title=title,labels=labels)
fig.show()
````
